#### Students Marks Analysis
The graphs indicate distribution of each subject marks. 

```python
plt.rcParams['figure.figsize'] = (20, 10)
sns.countplot(df_data['math score'])
plt.title('Math Score',fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
sns.countplot(df_data['reading score'], palette = 'Set1')
plt.title('Reading Score',fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
sns.countplot(df_data['writing score'], palette = 'Dark2')
plt.title('Writing Score',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(141)
plt.title('Math Scores')
sns.violinplot(y='math score',data=df_data,linewidth=2)
plt.subplot(142)
plt.title('Reading Scores')
sns.violinplot(y='reading score',data=df_data,linewidth=2)
plt.subplot(143)
plt.title('Writing Scores')
sns.violinplot(y='writing score',data=df_data,linewidth=2)
plt.show()
```

The maximum number of students have scored 60-80 in all the three subjects, math, reading and writing.

```python
from matplotlib.pyplot import xticks
plt.figure(figsize=(5,5))
sns.countplot(x="race/ethnicity",  data = df_data)
xticks(rotation=90)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Frequency')
plt.title('Race/Ethnicity')
```

The maximum number of students in each group are in group C, followed by group D, group B, group E and the least group A respectively.

```python
plt.subplot(141)
plt.title('Gender',fontsize = 20)
df_data['gender'].value_counts().plot.pie(autopct="%1.1f%%")
```

The list is dominated slightly by females.

```python
df_data["parental level of education"].unique()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x="parental level of education",hue = "gender", data= df_data, color='r')
plt.xlabel('Levels of Education',size='14')
plt.ylabel('Frequency')
plt.title('Parental Level of Education')
```

Highest proportion of parentel level of Education is 'Some college', 'associate's degreee' and 'high school'

```python
df_data["lunch"].unique()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x="lunch",hue = "gender", data= df_data)
plt.xlabel('Lunch Types')
plt.ylabel('Frequency')
plt.title('Lunch',size=14)
```

Most of the stodents have standard lunch preferences.

```python
plt.figure(figsize=(5,5))
sns.countplot(x="test preparation course",hue= "lunch", data= df_data, palette=['blue','c'])
plt.xlabel('preparation course')
plt.ylabel('Frequency')
```

Students who did not complete any perparation course and prefer standard lunch are the highest in number. It should be noted however that the number of students with no course and free lunch are nearly equal to the number of students with a completed course and standard lunch.

```python
plt.figure(figsize=(6,8))
sns.countplot(x="test preparation course",hue= "parental level of education", data= df_data, palette='Set1')
xticks(rotation=90)
plt.xlabel('Preparation Course')
plt.ylabel('Frequency')
```

Except in the high school case where there is a significant change between none and completed preparation course students, every other parental level of education maintains the same ratio in both none and coompleted courses.

```python
plt.figure(figsize=(15,4))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="gender", y="math score", data=df_data)
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(x="gender", y="reading score", data=df_data)
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(x="gender", y="writing score", data=df_data)
plt.show()
```

The male students scored higher in Maths whereas female students scored higher in Reading and writing.

```python
plt.figure(figsize=(15,6))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(131)
plt.title('Math Scores')
sns.barplot(x="race/ethnicity", y="math score", hue="test preparation course", data=df_data , palette = ['mediumaquamarine','coral'])
plt.subplot(132)
plt.title('Reading Scores')
sns.barplot(hue="test preparation course", y="reading score", x="race/ethnicity", data=df_data, palette = ['mediumaquamarine','coral'])
plt.subplot(133)
plt.title('Writing Scores')
sns.barplot(hue="test preparation course", y="writing score", x= 'race/ethnicity',data=df_data, palette = ['mediumaquamarine','coral'])

plt.show()
```

Highest number of Students who belongs to Group E has completed the test preperation course in Math and Reading and scored highest.
Highest number of Students who belongs to Group D and E has completed the test preperation course in Writing and scored highest.

```python
plt.figure(figsize=(16,5))
plt.subplot(1,3,1)
day_wise=df_data.groupby('race/ethnicity')['math score'].mean().plot.bar()
plt.title("Math Score")
plt.ylabel('average math score', fontsize=12)
plt.xlabel('race/ethnicity', fontsize=12)
plt.subplot(1,3,2)
day_wise=df_data.groupby('race/ethnicity')['reading score'].mean().plot.bar()
plt.title("Reading Score")
plt.ylabel('average reading score', fontsize=12)
plt.xlabel('race/ethnicity', fontsize=12)

plt.subplot(1,3,3)
day_wise=df_data.groupby('race/ethnicity')['writing score'].mean().plot.bar()
plt.title("Writing Score")
plt.ylabel('average writing score', fontsize=12)
plt.xlabel('race/ethnicity', fontsize=12)
```

In all the subject Group E dominates with the highest scores.

```python
x = pd.crosstab(df_data['race/ethnicity'], df_data['parental level of education'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar',
                                             stacked = 'True',
                                             figsize = (15, 8),
                                             color = plt.cm.winter(np.linspace(0, 1, 7)))


plt.title('Ethinicity vs Parental Education', fontsize = 20)
plt.grid()
plt.tight_layout()
plt.grid()
plt.show()
```

Very few parents have masters degree in all ethincities but Group D has the most number of parents with a masters degree.

```python
plt.figure(figsize=(40,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(252)
plt.title('Ethinicity and Lunch',fontsize=15)
sns.countplot(x="race/ethnicity", hue="lunch", data=df_data)
plt.show()
```

Group C has the highest number of students with standard lunch preferneces.

```python
sns.countplot(x = 'race/ethnicity', data = df_data,  hue = 'test preparation course', palette = 'Reds')
plt.title('Ethnicity vs Test Preparation', fontsize = 20)
plt.show()
```

Group C had the most number of completed test preparation courses.

```python
plt.figure(figsize=(15,5))
sns.lmplot(x="math score", y="reading score", hue="gender", data=df_data, markers=["o", "x"], palette="Set1")
plt.title('reading and math score with gender')

sns.lmplot(x="math score", y="writing score", hue="gender", data=df_data, markers=["o", "x"], palette="Set1")
plt.title('writing and  math score with gender')

sns.lmplot(x="reading score", y="writing score", hue="gender", data=df_data, markers=["o", "x"], palette="Set1")
plt.title('writing and reading score with gender')
```

Analysing each subject division graphs, writing and reading showed more uniformity whereas both writing and reading comparisions with math scores showed varied results. It can be deduced that reading and writing follow similar patterns.
